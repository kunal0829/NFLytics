```python
positionsDict = {
    'QB': 'QB',
    'RB': 'RB',
    'WR': 'WR',
    'TE': 'TE',
    
    'G': 'OG',
    'OG': 'OG',
    'T': 'OT',
    'OT': 'OT',
    'C': 'C',
    
    'DE': 'DE',
    'SS': 'S',
    'DB': 'CB',
    'CB': 'CB',
    'MLB': 'LB',
    'FS': 'S',
    'DT': 'DT',
    'OLB': 'LB',
    'ILB': 'LB',
    'NT': 'DT',
    'LB': 'LB',
    'SAF': 'S'
}

def getCorrectPos(posDict, pos):
    return posDict[pos]
```

```python
def returndf(location, position,teams,side):
    toreturn = pd.read_csv(location)
    toreturn = toreturn[(toreturn['Year']>=2013) & (toreturn['Year']<=2016) & (toreturn['Position'] == position)]
    toreturn = toreturn.filter(['Player Id','Year','Name','Position','Team'])
    toreturn['Player Id'] = toreturn['Player Id'].str.extract('(\d+)', expand=False).astype(int)
    toreturn = toreturn.replace({"Team": teams})
    toreturn = toreturn.replace({"Position": positionsDict})
    toreturn["Side"] = side
    toreturn.reset_index(drop=True, inplace=True)
    return toreturn
```

```python
def splitLine(location, positions,teams,side):
    toreturns = []
    for position in positions:
        toreturn = returndf(location, position, teams, side)
        toreturns.append(toreturn)
    ouput = pd.concat(toreturns, axis=0)
    ouput.reset_index(drop=True, inplace=True)
    return ouput
```

```python
offensiveLinePositions = ["G", 'C', 'OG', 'T', 'OT']
defensivePositions = ['DE', 'SS', 'DB', 'CB', 'MLB', 'FS', 'DT', 'OLB', 'ILB', 'NT', 'LB', 'SAF']

QBs = returndf("../newdata/Career_Stats_Passing.csv","QB", teamsDict, "Offense")
RBs = returndf("../newdata/Career_Stats_Rushing.csv","RB", teamsDict, "Offense")
WRs = returndf("../newdata/Career_Stats_Receiving.csv","WR",teamsDict, "Offense")
TEs = returndf("../newdata/Career_Stats_Receiving.csv","TE",teamsDict, "Offense")
OLs = splitLine("../newdata/Career_Stats_Offensive_Line.csv", offensiveLinePositions, teamsDict, "Offense")
Def = splitLine("../newdata/Career_Stats_Defensive.csv", defensivePositions, teamsDict, "Defense")
```

```python
Players = pd.concat([QBs, RBs, WRs, TEs, OLs, Def])
Players.reset_index(drop=True, inplace=True)
Names = Players["Name"].str.split(", ", n = 1, expand = True)
Players["First Name"] = Names[1]
Players["Last Name"] = Names[0]
Players.drop(['Name'], axis = 1, inplace = True)
order = ['Player Id', 'Year', 'First Name', 'Last Name', 'Team', 'Position', 'Side']
Players = Players[order]
Players.sort_values(by=['Year','Last Name', 'First Name'], inplace = True)
Players.reset_index(drop=True, inplace=True)
```
